```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# 1. Simulating our historical dataset (Features and Labels)
data = {
    'claim_amount': [1200, 8500, 400, 15000, 300, 9000, 1100, 2000],
    'vehicle_age': [2, 10, 1, 12, 3, 8, 4, 5],
    'history_of_accidents': [0, 2, 0, 3, 0, 1, 0, 1],
    'approved': [1, 0, 1, 0, 1, 0, 1, 1] # 1 = Approved, 0 = Rejected/Flagged
}
df = pd.DataFrame(data)

# 2. Separating Features (X) from the Target Label (y)
# X contains the inputs. y contains the answers.
X = df[['claim_amount', 'vehicle_age', 'history_of_accidents']]
y = df['approved']

# 3. The Train/Test Split
# We hold back 25% of the data to test the model like a final exam.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 4. Instantiate the Model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# 5. FIT (The Learning Phase)
# The model studies the relationships between X_train and y_train
model.fit(X_train, y_train)

# 6. PREDICT (The Exam Phase)
# We ask the model to predict outcomes for the test data it has never seen
predictions = model.predict(X_test)

print("Actual Outcomes:   ", list(y_test))
print("Model Predictions: ", list(predictions))
```
